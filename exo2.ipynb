```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

CA = pd.read_csv('CA entreprises.csv') # Source : https://ec.europa.eu/eurostat/databrowser/view/sts_setu_a/default/table?lang=fr&category=sts.sts_os.sts_os_t
protection_sociale = pd.read_csv('protection sociale.csv') # Source : https://ec.europa.eu/eurostat/databrowser/view/spr_net_ben/default/table?lang=fr
taux_chomage = pd.read_csv('chomage.csv') # Source : https://ec.europa.eu/eurostat/databrowser/view/une_educ_a/default/line?lang=fr&category=labour.employ.lfsi.une
CA.rename(columns={'TIME':'Pays / UE'}, inplace=True)
protection_sociale.rename(columns={'TIME':'Pays / UE'}, inplace=True)
taux_chomage.rename(columns={'TIME':'Pays / UE'}, inplace=True)
CA.rename(columns=lambda x: str(x), inplace=True)
protection_sociale.rename(columns=lambda x: str(x), inplace=True)
taux_chomage.rename(columns=lambda x: str(x), inplace=True)
CA.head()
```

```python
CA_France = CA[CA['Pays / UE'] == 'France'].iloc[0, 1:]
protection_sociale_France = protection_sociale[protection_sociale['Pays / UE'] == 'France'].iloc[0, 1:]
CA_ue_27 = CA[CA['Pays / UE'] == 'Union européenne - 27 pays (à partir de 2020)'].iloc[0, 1:]
protection_sociale_ue_27 = protection_sociale[protection_sociale['Pays / UE'] == 'Union européenne - 27 pays (à partir de 2020)'].iloc[0, 1:]

plt.plot(CA_France)
plt.xlabel('Années')
plt.ylabel('CA (rapporté en pourcentage au CA de 2021)')
plt.title('CA des entreprises du tertiaire en France selon les années')
plt.show()

plt.plot(protection_sociale_France)
plt.xlabel('Années')
plt.ylabel('Protection sociale (en millions d\'euros)')
plt.title('Protection sociale en France selon les années')
plt.show()

plt.plot(CA_ue_27)
plt.xlabel('Années')
plt.ylabel('CA (rapporté en pourcentage au CA de 2021)')
plt.title('CA des entreprises du tertiaire dans l\'UE-27 selon les années')
plt.show()

plt.plot(protection_sociale_ue_27)
plt.xlabel('Années')
plt.ylabel('Protection sociale (en millions d\'euros)')
plt.title('Protection sociale dans l\'UE-27 selon les années')
plt.show()
```

```python
taux_chomage.head()

taux_chomage_France = taux_chomage[taux_chomage['Pays / UE'] == 'France'].iloc[0, 1:]
taux_chomage_ue_27 = taux_chomage[taux_chomage['Pays / UE'] == 'Union européenne - 27 pays (à partir de 2020)'].iloc[0, 1:]

plt.plot(taux_chomage_France)
plt.xlabel('Années')
plt.ylabel('Taux de chômage')
plt.title('Taux de chômage en France selon les années')
plt.show()

plt.plot(taux_chomage_ue_27)
plt.xlabel('Années')
plt.ylabel('Taux de chômage')
plt.title('Taux de chômage dans l\'UE-27 selon les années')
plt.show()
```

```python
taux_chomage_France1 = [taux_chomage_France[k] for k in range( len(taux_chomage_France))]
CA_France1 = [CA_France[k] for k in range(len(CA_France))]
protection_sociale_France1 = [protection_sociale_France[k] for k in range(len(protection_sociale_France))]

plt.scatter(taux_chomage_France1[1:], CA_France1)
plt.xlabel('Taux de chômage')
plt.ylabel('CA (rapporté en pourcentage au CA de 2021)')
plt.title('CA des entreprises du tertiaire en France selon le taux de chômage')
plt.show()

plt.scatter(taux_chomage_France1[:-2], protection_sociale_France1[2:])
plt.xlabel('Taux de chômage')
plt.ylabel('Protection sociale (en millions d\'euros)')
plt.title('Protection sociale en France selon le taux de chômage')
plt.show()
```

```python
taux_chomage_Allemagne = taux_chomage[taux_chomage['Pays / UE'] == 'Allemagne'].iloc[0, 1:]
taux_chomage_Allemagne1 = [taux_chomage_Allemagne[k] for k in range(len(taux_chomage_Allemagne))]
CA_Allemagne = CA[CA['Pays / UE'] == 'Allemagne'].iloc[0, 1:]
CA_Allemagne1 = [CA_Allemagne[k] for k in range(1, len(CA_Allemagne))]
protection_sociale_Allemagne = protection_sociale[protection_sociale['Pays / UE'] == 'Allemagne'].iloc[0, 1:]
protection_sociale_Allemagne1 = [protection_sociale_Allemagne[k] for k in range(len(protection_sociale_Allemagne))]

plt.scatter(taux_chomage_Allemagne1[1:-1], CA_Allemagne1)
plt.xlabel('Taux de chômage')
plt.ylabel('CA (rapporté en pourcentage au CA de 2021)')
plt.title('CA des entreprises du tertiaire en Allemagne selon le taux de chômage')
plt.show()

plt.scatter(taux_chomage_Allemagne1[:-2], protection_sociale_Allemagne1[2:])
plt.xlabel('Taux de chômage')
plt.ylabel('Protection sociale (en millions d\'euros)')
plt.title('Protection sociale en Allemagne selon le taux de chômage')
plt.show()
```

```python
taux_chomage_ue_271 = [taux_chomage_ue_27[k] for k in range(len(taux_chomage_ue_27))]
CA_ue_271 = [CA_ue_27[k] for k in range(len(CA_ue_27))]
protection_sociale_ue_271 = [protection_sociale_ue_27[k] for k in range(len(protection_sociale_ue_27))]

plt.scatter(taux_chomage_ue_271[:-1], CA_ue_271)
plt.xlabel('Taux de chômage')
plt.ylabel('CA (rapporté en pourcentage au CA de 2021)')
plt.title('CA des entreprises du tertiaire dans l\'UE-27 en fonction du taux de chômage')
plt.show()

plt.scatter(taux_chomage_ue_271[:-2], protection_sociale_ue_271[2:])
plt.xlabel('Taux de chômage')
plt.ylabel('Protection sociale (en millions d\'euros)')
plt.title('Protection sociale dans l\'UE-27 en fonction du taux de chômage')
plt.show()
```
